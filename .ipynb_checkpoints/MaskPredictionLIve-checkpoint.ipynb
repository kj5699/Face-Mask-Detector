```python
import imutils
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import torch
import numpy as np
from torch import nn
from torch import optim
import os
from PIL import Image
import torch.nn.functional as F
from torchvision import datasets ,transforms, models
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from torch.autograd import Variable
```

```python
# transformation on Images
test_transforms=transforms.Compose([transforms.Resize((224,224)),transforms.ToTensor(),])

check_transforms=transforms.Compose([transforms.Resize((224,224)),transforms.ToTensor(),
                                     transforms.ToPILImage(),])

# check if gpu is available
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

#Load Model
model=torch.load('classifymodel_2.pth')
model.eval()
```

```python
def predict_image(image):
    image_tensor=test_transforms(image).float()
    image_tensor=image_tensor.unsqueeze_(0)
    input= Variable(image_tensor).to(device)
    output=model(input)
    
    index=output.data.cpu().numpy().argmax()
    return index

    
```

In this Notebook we Are using two different methods to Detect Face with opencv

1 Using HaarCascades
2 Using cv2.dnn method using pretrained Facenet Model 


```python
# using Haarcascasdes

def frontal_face(classifier,image,scaling_factor,classes):
    image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces=classifier.detectMultiScale(image_gray,scaling_factor,5)
    for x,y,w,h in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2)
        
        face= image[y:y+h,x:x+w]
        face=cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
        img = Image.fromarray(face)
        img=check_transforms(img)
        index=predict_image(img)
        label=str(classes[index])
        color = (0, 255, 0)
        cv2.putText(image, label, (x,y - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.45, color, 2)
        

    return image
```

```python
# using Cv2.dnn and Facenet
def detect_and_predict_mask(frame, facenet,classes):
    h,w=frame.shape[:2]
    blob=cv2.dnn.blobFromImage(frame,1.0,(300,300),(104.0,177.0,123.0))
    facenet.setInput(blob)
    detections=facenet.forward()
    
    
    
    for i in range(0,detections.shape[2]):
        confidence=detections[0,0,i,2]
        if confidence>0.5:
            box=detections[0,0,i,3:7]*np.array([w,h,w,h])
            (startX,startY,endX,endY)=box.astype("int")
            (startX, startY) = (max(0, startX), max(0, startY))
            (endX, endY) = (min(w - 1, endX), min(h - 1, endY))
            
            
            face = frame[startY:endY, startX:endX]
            face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(face)
            img=check_transforms(img)
            index=predict_image(img)
            label=str(classes[index])
            color = (0, 255, 0)
            cv2.putText(frame, label, (startX, startY - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.45, color, 2)
            cv2.rectangle(frame, (startX, startY), (endX, endY), color, 2)
            
    return frame
```

```python
# Uncomment follwing lines if you want to use haarcascades or lpb_face_cascades

#face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
#lpb_face_cascade=cv2.CascadeClassifier("lbpcascade_frontalface.xml")


prototxtPath="facedetectionmodels/deploy.prototxt"
weightsPath="facedetectionmodels/res10_300x300_ssd_iter_140000.caffemodel"
facenet = cv2.dnn.readNet(prototxtPath, weightsPath)

# classes to predict from
classes=['with mask','without mask']

    
```

```python
cap=cv2.VideoCapture(0)

while True:
    ret,image=cap.read()
    imutils.resize(image, width=200,height=200)
    """
    Uncomment next line and comment the succedding line to use haar 
    cascades for detecting face
    """
    
    #cv2.imshow("frame",frontal_face(face_cascade,image,1.2,classes)) 
    cv2.imshow("frame",detect_and_predict_mask(image, facenet,classes))
    k=cv2.waitKey(1) & 0xFF
    if k == ord("q"):
        print("Escapehit, closing...")
        break
cap.release()
cv2.destroyAllWindows()
```
